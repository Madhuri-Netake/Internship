### 9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description

```python
import selenium 
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains

import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time

warnings.filterwarnings('ignore')
driver = webdriver.Chrome()
driver.get("https://www.hostelworld.com")    #to open the Hostelworld website
time.sleep(2)

termAndConditons = driver.find_element(By.XPATH, '//button[@id="truste-consent-button"]')  # Accept the T and C
termAndConditons.click()


search = driver.find_element(By.XPATH, '/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[1]/div[1]/div/div[2]/input')  #to Insert 'sunglasses' in search bar
search.send_keys('London')

time.sleep(2)

search_button = driver.find_element(By.XPATH, '//ul[@class="select-list"]/li[2]')# Select the first element from list
search_button.click()

search_button = driver.find_element(By.XPATH, '//button[@class="btn-content large-button"]')# find the submit button and click
search_button.click()

time.sleep(7)


name=[]
property_rating =[]
distance =[]
total_reviews =[]
private_price= []
dorm_price = []
overall_rating = []

try:
    
    property_name = driver.find_elements(By.XPATH,'//div[@class="property-info-header"]/div/span')      #to scrap name of laptop names        
    for i in property_name:
        name.append(i.text)
     
    # find the rating element by using Xpath
    rating_elements = driver.find_elements(By.XPATH, '//div[@class="property-info-header"]//div[@class="property-rating"]/div//div[@class="rating-score"]/span[1]')
    for i in rating_elements :
        property_rating.append(i.text)    
    # Find the overall rating 
    overall_rating_elements = driver.find_elements(By.XPATH, '//div[@class="property-info-header"]//div[@class="property-rating"]/div//div[@class="rating-score"]/span[2]')
    for i in overall_rating_elements :
        overall_rating.append(i.text)    
    
    
    # Find the Distance element
    distance_elements = driver.find_elements(By.XPATH, '//div[@class="property-info-container"]//div[@class="property-distance"]/span[2]')
    for i in distance_elements :
        distance.append(i.text) 
        
    
     
    # Find the Display size and it's sibling element
    total_reviews_elements = driver.find_elements(By.XPATH, '//div[@class="property-info-header"]//div[@class="property-rating"]/div//div[@class="review"]/span[1]')
    for i in total_reviews_elements :
        total_reviews.append(i.text) 
        
    
    # Find the parent element 'property-accommodation-prices' so that we can check if 'No Privates Available' available in this 
    # element, If private 'No Privates Available' available in the element then add - to array, else add price to array
    private_price_elements = driver.find_elements(By.XPATH, '//div[@class="property-accommodation-prices"]/div[1]')
    for i in private_price_elements :
        price = i.text
        if price == "No Privates Available" :
            private_price.append("-")      
        else :
            p_price_element = i.find_elements(By.XPATH,'.//strong[@class="current"]')
            for j in p_price_element:
                private_price.append(j.text)
    
    # follow the same logic that we applied for private price 
    dorm_price_elements = driver.find_elements(By.XPATH, '//div[@class="property-accommodation-prices"]/div[2]')
    for i in dorm_price_elements :
        d_price = i.text
        if "No Dorms Available" in d_price :
            dorm_price.append("-")      
        else :
            d_price_element = i.find_elements(By.XPATH,'.//div[@class="accommodation-price single-row"]/strong')
            for j in d_price_element:
                dorm_price.append(j.text)              

finally:
    driver.close()

print(len(name),len(property_rating),len(overall_rating),len(distance),len(total_reviews),len(private_price),len(dorm_price))

df = pd.DataFrame({'Name':name,'Ratings':property_rating,'Overall Rating':overall_rating,'Distance':distance,'Total Reviews':total_reviews,'Private Price':private_price,'Dorm Price':dorm_price})
df
```
